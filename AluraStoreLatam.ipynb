# 2. Ventas por categor√≠a

```python
import pandas as pd


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


def productos_por_categoria(df, nombre_tienda):
    conteo = (
        df.groupby("Categor√≠a del Producto")["Producto"]
        .count()
        .reset_index()
        .sort_values(by="Producto", ascending=False)
    )
    conteo.columns = ["Categor√≠a", "Cantidad Vendida"]
    print(f"\nüìä Productos mas Populares en {nombre_tienda}:")
    print(conteo)
    return conteo

# Aplicar la funci√≥n a cada tienda
cat_t1 = productos_por_categoria(tienda1, "Tienda 1")
cat_t2 = productos_por_categoria(tienda2, "Tienda 2")
cat_t3 = productos_por_categoria(tienda3, "Tienda 3")
cat_t4 = productos_por_categoria(tienda4, "Tienda 4")
```

# 3. Calificaci√≥n promedio de la tienda


```python
import pandas as pd


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular calificaci√≥n promedio de cada tienda
promedios = {
    "Tienda 1": tienda1["Calificaci√≥n"].mean(),
    "Tienda 2": tienda2["Calificaci√≥n"].mean(),
    "Tienda 3": tienda3["Calificaci√≥n"].mean(),
    "Tienda 4": tienda4["Calificaci√≥n"].mean()
}


promedios_df = pd.DataFrame(list(promedios.items()), columns=["Tienda", "Calificaci√≥n Promedio"])

# Mostrar resultados con dos decimales
promedios_df["Calificaci√≥n Promedio"] = promedios_df["Calificaci√≥n Promedio"].round(2)
print(promedios_df)
```

# 4. Productos m√°s y menos vendidos

```python
import pandas as pd

# URLs de las bases de datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Funci√≥n para identificar los productos m√°s y menos vendidos
def productos_destacados(df, nombre_tienda):
    # Contar cantidad de ventas por producto
    conteo = (
        df["Producto"]
        .value_counts()
        .reset_index()
        .rename(columns={"index": "Producto", "Producto": "Cantidad Vendida"})
    )

    # Seleccionar los 5 m√°s y 5 menos vendidos
    top = conteo.head(5)
    bottom = conteo.tail(5)

    # Mostrar resultados en consola
    print(f"\nüè¨ {nombre_tienda} ‚Äî Productos m√°s vendidos:")
    print(top.to_string(index=False))

    print(f"\nüè¨ {nombre_tienda} ‚Äî Productos menos vendidos:")
    print(bottom.to_string(index=False))

    # Retornar resultados por si se desean usar despu√©s
    return top, bottom

# Aplicar la funci√≥n a cada tienda
top1, bottom1 = productos_destacados(tienda1, "Tienda 1")
top2, bottom2 = productos_destacados(tienda2, "Tienda 2")
top3, bottom3 = productos_destacados(tienda3, "Tienda 3")
top4, bottom4 = productos_destacados(tienda4, "Tienda 4")
```

# 5. Env√≠o promedio por tienda


```python
import pandas as pd

# URLs de las bases de datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos de cada tienda
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular el costo de env√≠o promedio
envio_promedio = {
    "Tienda 1": tienda1["Costo de env√≠o"].mean(),
    "Tienda 2": tienda2["Costo de env√≠o"].mean(),
    "Tienda 3": tienda3["Costo de env√≠o"].mean(),
    "Tienda 4": tienda4["Costo de env√≠o"].mean()
}

# Crear un DataFrame con los resultados
envio_promedio_df = pd.DataFrame(list(envio_promedio.items()), columns=["Tienda", "Costo de Env√≠o Promedio"])

# Redondear valores a 2 decimales para mejor lectura
envio_promedio_df["Costo de Env√≠o Promedio"] = envio_promedio_df["Costo de Env√≠o Promedio"].round(2)

# Mostrar los resultados
print("üì¶ Costo de env√≠o promedio por tienda (en COP):")
print(envio_promedio_df)
```

```python
# ============================================
# üìä AN√ÅLISIS FINAL Y GR√ÅFICOS COMPARATIVOS
# ============================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Cargar los datos desde las URLs oficiales ---
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# --- Calcular m√©tricas principales ---
ingresos = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

calificaciones = {
    "Tienda 1": tienda1["Calificaci√≥n"].mean(),
    "Tienda 2": tienda2["Calificaci√≥n"].mean(),
    "Tienda 3": tienda3["Calificaci√≥n"].mean(),
    "Tienda 4": tienda4["Calificaci√≥n"].mean()
}

envios = {
    "Tienda 1": tienda1["Costo de env√≠o"].mean(),
    "Tienda 2": tienda2["Costo de env√≠o"].mean(),
    "Tienda 3": tienda3["Costo de env√≠o"].mean(),
    "Tienda 4": tienda4["Costo de env√≠o"].mean()
}

# Configurar estilo general
sns.set(style="whitegrid", palette="Set2")

# =====================================================
# 1Ô∏è‚É£ Gr√°fico de BARRAS ‚Äì Ingresos Totales por Tienda
# =====================================================
plt.figure(figsize=(8,5))
plt.bar(ingresos.keys(), ingresos.values(), color=["#2E6DA4", "#5CB85C", "#F0AD4E", "#D9534F"])
plt.title("Ingresos Totales por Tienda", fontsize=14)
plt.xlabel("Tienda")
plt.ylabel("Ingreso Total (COP)")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()  # üëà asegura que se muestre

# =====================================================
# 2Ô∏è‚É£ Gr√°fico de DISPERSI√ìN ‚Äì Calificaci√≥n vs Costo de Env√≠o
# =====================================================
df_cal_envio = pd.DataFrame({
    "Tienda": list(calificaciones.keys()),
    "Calificaci√≥n Promedio": list(calificaciones.values()),
    "Costo de Env√≠o Promedio": list(envios.values())
})

plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df_cal_envio,
    x="Costo de Env√≠o Promedio",
    y="Calificaci√≥n Promedio",
    hue="Tienda",
    s=150,
    palette="Dark2"
)
plt.title("Relaci√≥n entre Calificaci√≥n y Costo de Env√≠o", fontsize=13)
plt.xlabel("Costo de Env√≠o Promedio (COP)")
plt.ylabel("Calificaci√≥n Promedio (1‚Äì5)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()  # üëà asegura que se muestre

# =====================================================
# 3Ô∏è‚É£ Gr√°fico CIRCULAR ‚Äì Distribuci√≥n de Categor√≠as (Tienda 2)
# =====================================================
if "Categor√≠a del Producto" in tienda2.columns:
    categorias_t2 = tienda2["Categor√≠a del Producto"].value_counts()

    plt.figure(figsize=(7,7))
    plt.pie(
        categorias_t2,
        labels=categorias_t2.index,
        autopct="%1.1f%%",
        startangle=90,
        colors=sns.color_palette("pastel"),
        wedgeprops={"edgecolor": "white"}
    )
    plt.title("Distribuci√≥n de Ventas por Categor√≠a ‚Äì Tienda 2", fontsize=14)
    plt.tight_layout()
    plt.show()  # üëà asegura que se muestre
else:
    print("‚ö†Ô∏è La columna 'Categor√≠a del Producto' no est√° disponible en Tienda 2.")

# =====================================================
# 4Ô∏è‚É£ Gr√°fico de L√çNEAS ‚Äì Comparaci√≥n de Calificaci√≥n Promedio
# =====================================================
plt.figure(figsize=(8,5))
plt.plot(
    calificaciones.keys(),
    calificaciones.values(),
    marker="o",
    color="#5CB85C",
    linewidth=2
)
plt.title("Comparaci√≥n de Calificaciones Promedio por Tienda", fontsize=14)
plt.xlabel("Tienda")
plt.ylabel("Calificaci√≥n Promedio (1‚Äì5)")
plt.ylim(0,5)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()  # üëà asegura que se muestre

# =====================================================
# üîç Conclusi√≥n visual general
# =====================================================
print("üìä CONCLUSI√ìN GENERAL:")
print("- Tienda 2 tiene los mayores ingresos y buena calificaci√≥n promedio.")
print("- Tienda 4 muestra menores ingresos y costos de env√≠o altos.")
print("- Tienda 2 domina las categor√≠as m√°s vendidas y presenta el mejor equilibrio general.")

```
